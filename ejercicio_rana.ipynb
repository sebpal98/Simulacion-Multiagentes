# Simulacion de computadores: La rana saltarina interdimensional

### Salto de la rana en 1 dimension

```python
from collections import Counter
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats


# Definir los puntos iniciales del paseo
x, y = [0], [0]

# Definir la longitud del paseo
n = 1000000

# Crear el paseo aleatorio
for i in range(n):
    # Elegir una direcciÃ³n aleatoria
    direccion = np.random.choice(["izquierda", "derecha"])
    
    # Mover la partÃ­cula una casilla en la direcciÃ³n elegida
    if direccion == "izquierda":
        x.append(x[-1] - 1)
        y.append(y[-1] - 1)
    elif direccion == "derecha":
        x.append(x[-1] + 1)
        y.append(y[-1] - 1)

# Crear una figura mÃ¡s grande y ajustar la escala
plt.figure(figsize=(12, 12), dpi=200)

# colorear los ejes en rojo
plt.axhline(y=0, color='r')
plt.axvline(x=0, color='r')

# Graficar el paseo aleatorio en 2D con cuadrÃ­cula
plt.plot(x, y)
plt.grid()

# Marcar el punto (0,0) en rojo
plt.plot(0, 0, 'ro')

#Graficar el punto final para saber donde llego la rana en su millonecimo salto
plt.plot(x[-1], y[-1], 'yo')

# Mostrar el grÃ¡fico y la Ãºltima coordenada
print("Ãšltima coordenada:", (x[-1], y[-1]))

# Mostrar el grÃ¡fico
plt.show()
```

```python
# Crear una lista de tuplas de (x, y)
coordenadas = list(x)

# Calcular el nÃºmero de elementos a mostrar
num_elementos = len(coordenadas) // 7000

# Tomar los primeros num_elementos elementos de la lista de coordenadas
coordenadas_a_mostrar = coordenadas[:num_elementos]

# Imprimir la lista de coordenadas en el formato deseado
for i, coord in enumerate(coordenadas_a_mostrar):
    if i > 0 and i % 5 == 0:
        print()
    print(coord, end=", \t\t")


# Lista de posiciones de la rana despuÃ©s de 1 millÃ³n de saltos
posiciones_rana = coordenadas

# Contamos cuÃ¡ntas veces la rana ha estado en cada posiciÃ³n
frecuencias = Counter(posiciones_rana)

# Convertimos los resultados en dos listas separadas
posiciones = list(frecuencias.keys())
frecuencias_posiciones = list(frecuencias.values())

# Crear una figura mÃ¡s grande y ajustar la escala
plt.figure(figsize=(8, 8), dpi=130)

# Graficamos las frecuencias
plt.bar(posiciones, frecuencias_posiciones)
plt.xlabel('Posiciones')
plt.ylabel('Frecuencias')

plt.show()



# Lista de posiciones de la rana despuÃ©s de 1 millÃ³n de saltos
posiciones_rana = coordenadas

# Realizamos el QQ plot
stats.probplot(posiciones_rana, dist='norm', plot=plt)

# Mostramos el grÃ¡fico
plt.title('QQ plot para las posiciones de la rana')
plt.show()
```

### Salto de la rana en 2 dimensiones

```python

# Definir los puntos iniciales del paseo
x, y = [0], [0]

# Definir la longitud del paseo
n = 1000000

# Crear el paseo aleatorio
for i in range(n):
    # Elegir una direcciÃ³n aleatoria
    direccion = np.random.choice(["arriba", "abajo", "izquierda", "derecha"])
    
    # Mover la partÃ­cula una casilla en la direcciÃ³n elegida
    if direccion == "arriba":
        x.append(x[-1])
        y.append(y[-1] + 1)
    elif direccion == "abajo":
        x.append(x[-1])
        y.append(y[-1] - 1)
    elif direccion == "izquierda":
        x.append(x[-1] - 1)
        y.append(y[-1])
    elif direccion == "derecha":
        x.append(x[-1] + 1)
        y.append(y[-1])
    if x[-1] == 250 and y[-1] == 300:
        print("La rana ha llegado a la posiciÃ³n (250, 300) en {} saltos.".format(i+1))
        break
    

# Crear una figura mÃ¡s grande y ajustar la escala
plt.figure(figsize=(12, 12), dpi=130)

# colorear los ejes en rojo
plt.axhline(y=0, color='r')
plt.axvline(x=0, color='r')

# Graficar el paseo aleatorio en 2D con cuadrÃ­cula
plt.plot(x, y)
plt.grid()

# Marcar el punto (0,0) en rojo
plt.plot(0, 0, 'ro')

#Graficar el punto final para saber donde llego la rana en su millonecimo salto
plt.plot(x[-1], y[-1], 'yo')

# Mostrar el grÃ¡fico y la Ãºltima coordenada
ultima_coordenada = (x[-1], y[-1])
print("Ãšltima coordenada:", ultima_coordenada)

# Mostrar el grÃ¡fico
plt.show()
```

#### Mostrar la lista de los datos

```python
# Crear una lista de tuplas de (x, y)
coordenadas = list(zip(x, y))

# Calcular el nÃºmero de elementos a mostrar
num_elementos = len(coordenadas) // 7000

# Tomar los primeros num_elementos elementos de la lista de coordenadas
coordenadas_a_mostrar = coordenadas[:num_elementos]

# Imprimir la lista de coordenadas en el formato deseado
for i, coord in enumerate(coordenadas_a_mostrar):
    if i > 0 and i % 5 == 0:
        print()
    print(coord, end=", \t\t")
```

### Salto de la rana en 3 dimensiones

```python

# Configurar la figura y el eje 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Definir los puntos iniciales del paseo
x, y, z = [0], [0], [0]

# Definir la longitud del paseo
n = 1000000

# Crear el paseo aleatorio
for i in range(n):
    # Elegir una direcciÃ³n aleatoria
    direccion = np.random.choice(["arriba", "abajo", "izquierda", "derecha", "adelante", "atras"])
    
    # Mover la partÃ­cula una casilla en la direcciÃ³n elegida
    if direccion == "arriba":
        x.append(x[-1])
        y.append(y[-1] + 1)
        z.append(z[-1])
    elif direccion == "abajo":
        x.append(x[-1])
        y.append(y[-1] - 1)
        z.append(z[-1])
    elif direccion == "izquierda":
        x.append(x[-1] - 1)
        y.append(y[-1])
        z.append(z[-1])
    elif direccion == "derecha":
        x.append(x[-1] + 1)
        y.append(y[-1])
        z.append(z[-1])
    elif direccion == "adelante":
        x.append(x[-1])
        y.append(y[-1])
        z.append(z[-1] + 1)
    elif direccion == "atras":
        x.append(x[-1])
        y.append(y[-1])
        z.append(z[-1] - 1)
    if x[-1] == 45 and y[-1] == 23 and z[-1] == 17:
        print("La rana ha llegado a la posiciÃ³n (45, 23, 17) en {} saltos.".format(i+1))
        break

# Graficar el paseo aleatorio en 3D
plt.plot(0, 0, 0, 'ro')
ax.plot(x, y, z)

# Marcar el punto (0,0) en rojo
plt.plot(0, 0, 0, 'ro')

# Graficar el punto final para saber donde llego la rana en su millonecimo salto
plt.plot(x[-1], y[-1], z[-1], 'yo')

# Cambiar el tamaÃ±o de la figura
fig.set_size_inches(12, 12)

# Mostrar el grÃ¡fico y la Ãºltima coordenada
ultima_coordenada = (x[-1], y[-1], z[-1])
print("Ãšltima coordenada:", ultima_coordenada)

# Mostrar el grÃ¡fico
plt.show()
```

#### Mostrar la lista de los datos

```python
# Crear una lista de tuplas de (x, y, z)
coordenadas = list(zip(x, y, z))

# Calcular el nÃºmero de elementos a mostrar
num_elementos = len(coordenadas) // 7000

# Tomar los primeros num_elementos elementos de la lista de coordenadas
coordenadas_a_mostrar = coordenadas[:num_elementos]

# Imprimir la lista de coordenadas en el formato deseado
for i, coord in enumerate(coordenadas_a_mostrar):
    if i > 0 and i % 5 == 0:
        print()
    print(coord, end=", \t\t")
```
